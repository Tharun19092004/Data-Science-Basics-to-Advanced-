```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import sigmoid_kernel,linear_kernel,polynomial_kernel,rbf_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df= pd.read_csv('netfilx_datasets.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df.description
```

```python
tfidf=TfidfVectorizer(stop_words='english')
```

```python
tfidf1=tfidf.fit_transform(df.description)
```

```python
tfidf1
```

```python
similarities=sigmoid_kernel(tfidf1,tfidf1)
similarities[1]
```

```python
similarities.shape
```

```python
indices= pd.Series(data=df.index,index=df.title)
indices
```

```python
def recommended_movie(similar_movie):
    if similar_movie in indices.index:
        index= np.where(similar_movie==indices.index)[0][0]
        similar=sorted(list(enumerate(similarities[index])),reverse=True,key=lambda x: x[1])[1:6]
        print(f'Recommended movie of {similar_movie}')
        print('*'*30)
        for movie in similar:
            print(indices.index[movie[0]])
    else:
        print('Movie is not in the list')
```

```python
recommended_movie('Zodiac')
```
