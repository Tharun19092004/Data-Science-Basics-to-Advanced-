```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('driver-data.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.head()
```

```python
features= df.drop(columns=['id'])
```

```python
std_sca=StandardScaler()
features= pd.DataFrame(std_sca.fit_transform(features),columns=features.columns)
features.head()
```

```python
kmeans= KMeans(n_clusters=3)
cluster= kmeans.fit(features)
np.unique(cluster.labels_)
```

```python
df['cluster']= cluster.labels_
df[df.cluster==1]
```

```python
sns.lmplot(data=df, x='mean_dist_day',y='mean_over_speed_perc',hue='cluster')
```

```python
silhouette_score(features,cluster.labels_)
```

```python
kmeans.inertia_
```

#### Elbow Method

```python
inertia=[]
for i in range(1,11):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
print(inertia)
```

```python
plt.plot(range(1,11) ,inertia)
plt.xlabel('N_of_clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
```

```python
kmeans= KMeans(n_clusters=4)
cluster= kmeans.fit(features)
np.unique(cluster.labels_)
```

```python
silhouette_score(features,cluster.labels_)
```
